**Step 5: Data Visualization and Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visualisation style
sns.set_style('whitegrid')

# Visualize the distribution of price ranges
plt.figure(figsize=(10, 6))
sns.countplot(x='price_range', data=train_data)
plt.title('Distribution of Price Ranges')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()

# Visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(train_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Visualize the impact of battery power on price range
plt.figure(figsize=(10, 6))
sns.boxplot(x='price_range', y='battery_power', data=train_data)
plt.title('Battery Power vs. Price Range')
plt.xlabel('Price Range')
plt.ylabel('Battery Power')
plt.show()

# Visualize the impact of RAM on price range
plt.figure(figsize=(10, 6))
sns.boxplot(x='price_range', y='ram', data=train_data)
plt.title('RAM vs. Price Range')
plt.xlabel('Price Range')
plt.ylabel('RAM')
plt.show()

# Visualize the impact of pixel resolution on price range
plt.figure(figsize=(10, 6))
sns.boxplot(x='price_range', y='px_height', data=train_data)
plt.title('Pixel Height vs. Price Range')
plt.xlabel('Price Range')
plt.ylabel('Pixel Height')
plt.show()
```

**Insights**

From the visualizations, we can derive several insights:

**Distribution of Price Ranges:** The price ranges are evenly distributed, indicating a balanced dataset.

**Correlation Matrix:** The heatmap shows the correlations between different features, highlighting which factors have strong relationships with the target variable (price_range). For example, RAM has a strong positive correlation with the price range.

**Battery Power vs. Price Range:** Higher battery power tends to be associated with higher price ranges.

**RAM vs. Price Range:** Higher RAM is strongly associated with higher price ranges.

**Pixel Height vs. Price Range:** Higher pixel resolution (height) is associated with higher price ranges.
